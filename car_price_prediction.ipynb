```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('car_price.csv')
df = df.rename(columns = {'Unnamed: 0':'Car_ID'})
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
def convert_to_rupees(price):
    if 'Lakh' in price:
        return float(price.replace(' Lakh', '').replace(',', '')) * 1_00_000
    elif 'Crore' in price:
        return float(price.replace(' Crore', '').replace(',', '')) * 1_00_00_000
    else:
        return float(price.replace(',', ''))
```

```python
df.columns = [col.strip() for col in df.columns] 
df['kms_driven'] = df['kms_driven'].str.replace(' kms', '').str.replace(',', '').astype(float)
df['Seats'] = df['Seats'].str.replace(' Seats', '').astype(float)
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].apply(convert_to_rupees)
```

```python
predict = "car_prices_in_rupee"
features = ["Car_ID", "car_name", "kms_driven", "fuel_type", "transmission", "ownership", "manufacture", "engine", "Seats"]
```

```python
data = df[features + [predict]]
data = data.dropna()
```

```python
X = data[features]
y = data[predict]
```

```python
X = pd.get_dummies(X, columns=["car_name", "kms_driven", "fuel_type", "transmission", "ownership", "manufacture", "engine", "Seats"])

X = X.values
y = y.values
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=32)
```

```python
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
score = model.score(xtest, ytest)
```

```python
print(f"Model Score: {score * 100}")
```
